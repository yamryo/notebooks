# 3次関数のグラフと接線、および関連する領域の面積

```python
from sympy import *
init_printing()

from sympy.plotting import plot
```

```python
x,y = symbols('x y')
```

```python
f=lambdify(x, x*(x**2-1))
f(0)
```

```python
df = lambda x: 3*x**2-1 #diff(f(x))
df(0)
```

```python
a = 1/2
g = lambda x: df(a)*(x-a)+f(a)
# g(x)

p1 = plot(f(x), (x,-2,2), ylim=(-2,2), show=False)
p2 = plot(g(x), (x,-2,2), ylim=(-2,2), show=False)
p1.append(p2[0])

p1.show()
```

```python
sol = solveset(f(x)-g(x), x)
itg = Integral(f(x)-g(x),(x,min(sol),max(sol)))
itg, itg.doit()
```

```python
s = lambda t: abs(Integral(f(t)-g(t))
p3 = plot(s(x), (x,-2,2), ylim=(-2,2), show=False)
p3.show()
```

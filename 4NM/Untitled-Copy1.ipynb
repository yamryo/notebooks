```ruby
require 'rbplotly'
```

```ruby
trace0 = { x: [0,1], y: [1,0], mode: :markers}

N = 1000
px = (N+1).times.map{|t| rand(0..10.0)*((rand(0..1)==0) ? 1 : -1)}
py = (N+1).times.map{|t| rand(0..10.0)*((rand(0..1)==0) ? 1 : -1)}
tP = { x: px, y: py, mode: :markers }

gx = px.map{|x| (0+1+x).to_f/3 }
gy = py.map{|y| (0+0+y).to_f/3 }
tG = { x: gx, y: gy, mode: :markers }

trace = [trace0, tP, tG]

layout = {height: 800, width: 800}
plot = Plotly::Plot.new(data: trace, layout: layout)
plot.show
nil
```

```ruby
trace0 = { x: [1,0], y: [0,1], mode: :markers}

N = 10**5
pxa, pya, gxa, gya = [], [], [], []
(N+1).times.each do |k|
  px = rand(0..1.0)*((rand(0..1)==0) ? 1 : -1)
  py = rand(0..1.0)*((rand(0..1)==0) ? 1 : -1)
  gx, gy = (1+0+px).to_f/3, (0+1+py).to_f/3
  
  if (((px-1)**2+py**2) - gx**2).abs < 0.001
    pxa << px; pya << py
    gxa << gx; gya << gy
  end
end

tP = { x: pxa, y: pya, mode: :markers }
tG = { x: gxa, y: gya, mode: :markers }
trace = [trace0, tP, tG]

layout = {height: 800, width: 780}
plot = Plotly::Plot.new(data: trace, layout: layout)
plot.show
nil
```

```ruby
pA, pB = [2, 0], [0, 2]
trace0 = { x: [pA[0], pB[0]], y: [pA[1], pB[1]], mode: :'markers+lines'}

N = 10**5
pxa, pya, gxa, gya = [], [], [], []
(N+1).times.each do |k|
  px = rand(0..pA[0].to_f)*((rand(0..1)==0) ? 1 : -1)
  py = rand(0..pB[1].to_f)*((rand(0..1)==0) ? 1 : -1)
  gx, gy = (pA[0]+pB[0]+px).to_f/3, (pA[1]+pB[1]+py).to_f/3
  
  if ((px**2+py**2) - (gx**2+0*gy**2)).abs < 0.001
    pxa << px; pya << py
    gxa << gx; gya << gy
  end
end

tP = { x: pxa, y: pya, mode: :markers, marker: {size: 3} }
tG = { x: gxa, y: gya, mode: :markers, marker: {size: 3} }
tL = pxa.size.times.map do |i|
  {x: [pxa[i], gxa[i]], y: [pya[i], gya[i]], mode: :lines, line: {color: 'rgb(125,125,125)', width: 1} }
end
trace = [trace0, tP, tG] + tL

layout = {height: 800, width: 800}
plot = Plotly::Plot.new(data: trace, layout: layout)
plot.show
nil
```

```ruby
N = 50
x, y = [], []
en = (N+1).times.map do |i|
  x << Math.cos(2*Math::PI*(i.to_f/N.to_f))
  y << Math.sin(2*Math::PI*(i.to_f/N.to_f))
end
circle = { x: x, y: y, mode: :lines, line: {color: 'rgb(125,125,125)', width: 1} }
line1 = {x: [-1,3], y: [1,1], mode: lines}
line2 = {x: [-1,3], y: [-1,-1], mode: lines}

base = [circle, line1, line2]

n = 6
lines = n.times.map do |k|
  t = (3*(k+1)/n.to_f)
  pP, pQ = [t,1], [1/t, -1]
  { x: [pP[0], pQ[0]], y: [pP[1], pQ[1]], mode: :'markers+lines', marker: {size: 12} }
end

trace = base + lines 

layout = {height: 600, width: 1000}
plot = Plotly::Plot.new(data: trace, layout: layout)
plot.show
nil
```

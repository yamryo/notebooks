```python
from __future__ import print_function, division
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.path as mpath
import matplotlib.patches as mpatches
%matplotlib inline

import math
import sympy
```

```python
fig, ax = plt.subplots(figsize=(10,10))

N = 100
circle = [(k/N)*2*np.pi for k in range(N)]
X = [(1+ np.sqrt(2))*np.cos(t) for t in circle]
Y = [np.sin(t) for t in circle]

plt.plot()

ax.grid()
ax.axis('equal')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

N = 100
circle = [(k/N)*2*np.pi for k in range(N)]
X = [np.sqrt(2)*(1+np.cos(t)) for t in circle]
Y = [1+np.sqrt(2)*np.sin(t) for t in circle]

Path = mpath.Path
path_data = []
for k in range(N):
    path_data.append((Path.MOVETO, (X[k], Y[k])))
codes, verts = zip(*path_data)
path = mpath.Path(verts, codes)
patch = mpatches.PathPatch(path, facecolor='r', alpha=0.5)
ax.add_patch(patch)

x, y = zip(*path.vertices)
line, = ax.plot(x, y, 'b.', ms = 3.0)

ax.grid()
ax.axis('equal')
plt.show()
```

# The number of zeros in $n!$

```python
from __future__ import print_function, division
import itertools
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.path as mpath
import matplotlib.patches as mpatches
%matplotlib inline

import math
```

```python
dig='0'
# marr = [str(math.factorial(k)).count(dig) for k in range(100)]
# print(marr)

fig, ax = plt.subplots(figsize=(18,18))
N = 5000

X = list(range(N))
Y = [str(math.factorial(x)).count(dig) for x in X]
Z = [int(x/5) for x in X]

ax.plot(X, Y, 'b.', ms = 3.0)
ax.plot(X, Z, 'r.', ms = 3.0)
# plot1 = ax.plot(X, Y, 'b.', ms = 3.0)
# plot2 = ax.plot(X, Z, 'r.', ms = 3.0)

ax.grid()
# ax.axis('equal')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18,18))
N = 1000

X = list(range(N))
Y = [0] * 10
color_cycle= itertools.cycle(["orange","pink","blue","brown","red","grey","yellow","green","azure","plum"])

for s in list(range(10)):
    Y[s] = [str(math.factorial(x)).count(str(s)) for x in X]
    ax.plot(X, Y[s], '.', color=next(color_cycle), ms = 3.0)    
    
ax.grid()
# ax.axis('equal')
plt.show()
```

```python
N = 1000
X = list(range(N))

Row, Col = 5, 2
fig = plt.figure(figsize=(10,20))
ax = [0]*(Row*Col)

# ---
for s in range(Row*Col):    
    Y = [str(math.factorial(x)).count(str(s)) for x in X]
    ax[s] = fig.add_subplot(Row, Col, s+1)
    ax[s].scatter(X,Y, s=1.0)
#     ax[s].set_xlim(0,200)
    ax[s].set_ylim(0,int(N/2))
    ax[s].set_title("dig={}".format(s))
    ax[s].grid()
# ---
fig.tight_layout()
fig.show()
```

# Curvature and Torsion of a Space Curve

```python
from sympy import *
from sympy.plotting import plot3d_parametric_line
init_printing(use_unicode=True)
from IPython.display import display, Math
```

# 1. A Space Curve

```python
t = symbols("t")

p = [t,t**2,t**3] #[5*cos(t), 5*sin(t), 2*t] #[t, t**2/2, t**4/4] #[t,t**3/3,t**4/4] #[t, Rational(1,2)*t**2, Rational(1,3)*t**3] #

plot3d_parametric_line(
    p[0],p[1],p[2],
    (t, -5, 5)
)
```

# 2. Curvature and Torsion

```python
dp = [f.diff(t,1).ratsimp() for f in p]
IdpI = sqrt(sum([k**2 for k in dp]).factor())

e_1 = [k/IdpI for k in dp]
display('e_1=',e_1)
```

```python
ddp = [((f.diff(t,1))*(1/IdpI)).ratsimp() for f in e_1]
display('(d/ds)e_1=', ddp)
        
ddp_int = [(f*(IdpI**4)).ratsimp() for f in ddp]
# ddp_int
IddpI = sqrt(sum([f**2 for f in ddp_int]).factor())
# ddpr.ratsimp()
e_2 = [f/IddpI for f in ddp_int]
display('e_2=', e_2)
```

```python
kappa = (ddp[0]/e_2[0]).factor().simplify()
display('kappa=', kappa)
```

```python
# e_3nn = Matrix(e_1).cross(Matrix(e_2))
e_3nn = Matrix(dp).cross(Matrix(ddp_int))

fac = 1 #((IddpI/IdpI)**2).simplify()
e_3nn = [(f/(IdpI**2)/sqrt(fac)).simplify().cancel() for f in e_3nn]
e_3nn
Ie_3I = sqrt(sum([f**2 for f in e_3nn]).factor())
e_3 = [f/Ie_3I for f in e_3nn]
display('e_3=', e_3)
```

```python
de_3 = [f.diff(t,1)*(1/IdpI).ratsimp() for f in e_3]
# display('(d/ds)e_3=', de_3)

tau=(-de_3[0]/e_2[0]).simplify().factor()
display('tau=', tau)
```

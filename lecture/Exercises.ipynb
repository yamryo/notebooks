## Colatz problem

1. Write a function, say `collatz()`, defined as follows:
\begin{equation}
  \forall n \in \mathbb{N},~~~~
  collatz(n) = \begin{cases}
  0 & n=1 \\
  \dfrac{n}{2} & n:\text{even} \\
  3n+1 & \text{otherwise}
  \end{cases}
\end{equation}

1. Confirm the Collatz conjecture for $n<100$.
> **Conjecture**. For $\forall n \in \mathbb{N}$, the sequence generated by iteration of collatz function starting from $n$ always reach $1$.

1. Write a function `collatz_list()` to generate a list of numbers from the given number to $1$ by iteration of the collatz function.

1. Plot a graph of the function `len(collatz_list(n))`.

> **Knowledge**: `int`, `list`, `def function():`, `if-else`, `while`

## Prime numbers

1. Create a function `is_prime()` which determine that a given number is a prime or not.

1. Get a list prime numbers up to $10000$.

1. Confirm the prime theorem.
> **The Prime Theorem**. 
\begin{equation}
  \pi(n) \sim \frac{n}{\log{n}} ~~~~
  (x \to \infty),
\end{equation}
where $\pi(n)$ is the number of primes up to $n \in \mathbb{N}$.

1. Find the first prime number $p$ which the Mersenne type number $2^p-1$ of $p$ is not prime.

1. Get a list of Mersenne primes up to 2^40.

1. Confirm the following theorem for all perfect numbers up to $10000$.

> **Theorem**(L. Euler). 
The all even perfect numbers can be written as $2^{p-1}(2^{p}-1)$ with a prime $p$.

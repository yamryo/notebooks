# Kaprekar Numbers

## Define Kaprekar Class

```ruby
module Kaprekar_interface
  def get_next() end
  def sequence() end
  def find_loop() end
  
  def ==(kp) @number == kp.number end
  def inspect() @number end

  protected
  def split(num)
    raise ArgumentError unless num.is_a?(Integer)
    return num.to_s.split("")
  end
  def min(num) split(num).sort.join.to_i end
  def max(num) split(num).sort.reverse.join.to_i end
end

#-------------------------------------------------------#

class Kaprekar include Kaprekar_interface
  def initialize(num)
    @number = num
    @next = max(num) - min(num)
  end
  attr_reader :number, :next, :prevs
  
  def get_prevs()
    dgt = 10**(@number.to_s.size-1)
    rng = (dgt..dgt*10)
    return rng.to_a.select{|n| self.class.new(n).next == @number}
  end
  
  def sequence(count=200)
    seq = [self]
    while (seq.count(seq[-1]) < 2 and count > 0)
      seq << self.class.new(seq[-1].next)
      count -= 1
    end
    seq.pop 
    return seq
  end

  def find_loop()
    seq = self.sequence
    idx = seq.index(seq[-1])
    return seq[idx, seq.length - idx]
  end
  
  def to_i()
    return @number
  end
  def to_s()
    return @number.to_s
  end
end; nil
```

```ruby
k1 = Kaprekar.new(198)
[k1,k1.next, k1.get_prevs]
```

```ruby
kp = Kaprekar.new(102)
[kp.sequence, kp.find_loop]
```

```ruby
nn = 10**2
(nn..nn*10).each_with_object([]) do |n, a|
  kp = Kaprekar.new(n)
  mystr = kp.number.to_s
  while !a.include?(kp)
    a << kp
    kp = Kaprekar.new(kp.next)
    mystr += '->'+ kp.number.to_s
  end
  p mystr
end; nil
```

## Display Kaprekar Tree

```ruby
class KpTree
  def initialize(kp, maxdepth=10)
    @root = kp
    @subtree = self.maketree(maxdepth)
  end
  
  def show(stack=[], indent=0, keisen='├─ ')
    branch = (indent==0) ? '' : ' '*(2+5*(indent-1)) + keisen
    unless stack.include?(@root.number)
      puts branch + @root.to_s
      stack << @root.number
      unless @subtree.nil?
        @subtree.each_with_index do |c, idx|
          keisen = '└─ ' if idx==@subtree.size-1
          c.show(stack, indent+1, keisen)
        end
      end
    else
      puts branch + "(#{@root.to_s})"
    end
    return stack
  end
  
  protected
  def maketree(maxdepth=10)
    children = @root.get_prevs
    unless (children.empty? or maxdepth==0) then
      subtree = children.map{|c| self.class.new(Kaprekar.new(c), maxdepth-1) }
    end
    return subtree
  end
end; nil
```
